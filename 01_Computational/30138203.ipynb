```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Import CSV load dataset
h_d = '/content/drive/MyDrive/Colab Notebooks/healthcare_dataset.csv'
data = pd.read_csv(h_d)

# Display the first few rows.
print(data.head())
```

```python
# Display the Last few rows.
print(data.tail())
```

```python
# Rename Columns in dataset
data.rename(columns={'Medication':'Medication_Prescribed','Id':'Paitent_ID'})
```

```python
# Display the dataset info
print(data.info())
```

```python
# Check for missing values in the dataset
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values[missing_values > 0])
```

```python
# Calculate the missing data in percentage
data.isnull().sum()/data.shape[0]*100
```

```python
# Convert the Name column to proper case.
data['Name'] = data['Name'].str.title()
```

```python
# Convert Id column from object to float.
data['Id'] = pd.to_numeric(data['Id'], errors='coerce')
```

```python
# Convert Date of Admission and Discharge Date to datetime format.
data['Date of Admission'] = pd.to_datetime(data['Date of Admission'], errors='coerce', dayfirst=True)
data['Discharge Date'] = pd.to_datetime(data['Discharge Date'], errors='coerce', dayfirst=True)
```

```python
# Convert columns with ambiguous data types.
data['Billing Amount'] = pd.to_numeric(data['Billing Amount'], errors='coerce')
```

```python
# Drop columns with high missing values if deemed irrelevant.
data.drop(columns=['BloodPressure', 'SkinThickness', 'Insulin', 'Id'], inplace=True)
```

```python
# Drop Duplicates from dataset
data.drop_duplicates(inplace=True)
```

```python
# Fill NaN values in age and room_number columns with their respective means.
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Room Number'].fillna(data['Room Number'].mean(), inplace=True)

# Convert 'age' and 'room_number' columns from float to integer
data['Age'] = data['Age'].astype(int)
data['Room Number'] = data['Room Number'].astype(int)
```

```python
# Fill NaN values in Age and Room Number columns with their respective means
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Room Number'].fillna(data['Room Number'].mean(), inplace=True)

if 'Billing Amount' in data.columns:
    data['Billing Amount'].fillna(data['Billing Amount'].mean(), inplace=True)
```

```python
# For categorical columns, fill NaNs with the most frequent value (mode)
categorical_columns = ['Gender', 'Blood Type', 'Medical Condition', 'Doctor', 'Hospital',
                       'Insurance Provider', 'Test Results', 'Neighbourhood', 'Age Group']
for col in categorical_columns:
    if col in data.columns:
        if not data[col].dropna().empty:
            mode_value = data[col].mode()[0]
            data[col].fillna(mode_value, inplace=True)
```

```python
# Fill missing values in the Name column with random names with mode
mode_name = data['Name'].mode()[0]
data['Name'].fillna(mode_name, inplace=True)
```

```python
# Replace missing dates with a random date
start_date = pd.to_datetime("2022-01-01")
end_date = pd.to_datetime("2023-12-31")

date_range = pd.date_range(start=start_date, end=end_date)

data['Date of Admission'] = data['Date of Admission'].apply(
    lambda x: np.random.choice(date_range) if pd.isna(x) else x
)
data['Discharge Date'] = data['Discharge Date'].apply(
    lambda x: np.random.choice(date_range) if pd.isna(x) else x
)
```

```python
# Fill missing values in Admission Type with the mode (most frequent value)
if 'Admission Type' in data.columns:
    mode_admission_type = data['Admission Type'].mode()[0]
    data['Admission Type'].fillna(mode_admission_type, inplace=True)
```

```python
# Fill missing values in Medication with Not Specified
data['Medication'].fillna("Not Specified", inplace=True)
```

```python
# Columns to fill missing values in
columns_to_fill = ['Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received', 'Showed_up']

# Fill missing values with False in the specified columns
data[columns_to_fill] = data[columns_to_fill].fillna(False)
```

```python
# Transforming age into age groups for better insights
bins = [0, 18, 35, 50, 65, np.inf]
labels = ['Child', 'Young Adult', 'Adult', 'Middle Aged', 'Senior']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels)
```

```python
# Outlier Detection and Handling in Billing Amount
upper_limit = data['Billing Amount'].quantile(0.99)
data['Billing Amount'] = np.where(data['Billing Amount'] > upper_limit, upper_limit, data['Billing Amount'])
```

```python
# Importing necessary libraries
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from scipy.stats import linregress
```

```python
# Converting categorical variables like Gender and Blood Type to numerical values.
label_cols = ['Gender', 'Blood Type', 'Medical Condition', 'Doctor', 'Hospital', 'Insurance Provider', 'Neighbourhood']
label_encoders = {}

for col in label_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))
    label_encoders[col] = le
```

```python
# Define age groups (e.g.Child,Young Adult, etc.)
age_bins = [0, 18, 35, 50, 65, np.inf]
age_labels = ['Child', 'Young Adult', 'Adult', 'Middle Aged', 'Senior']
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)
```

```python
# Convert Date of Admission to datetime and create a new feature for days since admission.
from datetime import datetime
data['Date of Admission'] = pd.to_datetime(data['Date of Admission'], errors='coerce',dayfirst=True)
today=pd.to_datetime(datetime.today().date())
data['Days Since Admission'] = (pd.to_datetime('today') - data['Date of Admission']).dt.days
```

```python
# Create a new feature summarizing the count of conditions like Hipertension,Diabetes, etc.
condition_cols = ['Hipertension', 'Diabetes', 'Alcoholism', 'Handcap']
data['Total Conditions'] = data[condition_cols].sum(axis=1)
```

```python
# Flagging High Billing Amounts.
high_billing_threshold = data['Billing Amount'].quantile(0.75)
data['High Billing'] = data['Billing Amount'] > high_billing_threshold
```

```python
# Encode Gender into numeric format (e.g., Male=1, Female=0).
data['Gender_encoded'] = data['Gender'].apply(lambda x: 1 if x == 'Male' else 0)

# Encode Admission Type using one-hot encoding.
data = pd.get_dummies(data, columns=['Admission Type'], drop_first=True)
```

```python
# Binary Encoding for Attendance Outcome
# Convert 'Showed_up' to binary (1 for True, 0 for False)
data['Showed_up'] = data['Showed_up'].astype(int)
```

```python
# Age Distribution by Age Group
plt.figure(figsize=(10, 6))
sns.countplot(x='Age Group', data=data, palette='viridis', order=['Child', 'Young Adult', 'Adult', 'Middle Aged', 'Senior'])
plt.title("Distribution of Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Number of Patients")
plt.show()
```

```python
# Calculate Length of Stay in days
data['Length of Stay'] = (data['Discharge Date'] - data['Date of Admission']).dt.days
plt.figure(figsize=(10, 5))
plt.scatter(data['Age'], data['Length of Stay'], alpha=0.6)
plt.title("Scatter Plot of Length of Stay vs Age")
plt.xlabel("Age")
plt.ylabel("Length of Stay (Days)")
plt.show()
```

```python
# Total Billing Amount Over Time (by month)

data['Billing Amount'] = pd.to_numeric(data['Billing Amount'], errors='coerce')
billing_over_time = data.groupby(data['Date of Admission'].dt.to_period('M'))['Billing Amount'].sum()

plt.figure(figsize=(10, 6))
billing_over_time.plot(linewidth=2)
plt.title("Total Billing Amount Collected Over Time (Monthly)")
plt.xlabel("Date")
plt.ylabel("Total Billing Amount")
plt.show()
```

```python
# Plotting for age vs blood type
blood_age_data = data[['Age', 'Blood Type']].dropna()

plt.figure(figsize=(10, 6))
blood_age_data.boxplot(column='Age', by='Blood Type')
plt.title('Age Distribution by Blood Type')
plt.suptitle('')
plt.xlabel('Blood Type')
plt.ylabel('Age')
plt.xticks(rotation=10)
plt.grid(False)
plt.show()
```

```python
# High Billing Flag by Age Group
plt.figure(figsize=(10, 6))
sns.countplot(x='Age Group', hue='High Billing', data=data, palette='pastel')
plt.title("High Billing Status by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.legend(title="High Billing")
plt.show()
```

```python
# Regression Plot: Billing Amount vs. Length of Stay
plt.figure(figsize=(6, 4))
sns.regplot(x="Length of Stay", y="Billing Amount", data=data)
plt.title("Regression Analysis: Billing Amount vs Length of Stay")
plt.xlabel("Length of Stay (Days)")
plt.ylabel("Billing Amount ($)")
plt.show()
```

```python
# Regression Plot: Billing Amount vs. Total Conditions
plt.figure(figsize=(6, 4))
sns.regplot(x="Total Conditions", y="Billing Amount", data=data)
plt.title("Regression Analysis: Billing Amount vs Total Conditions")
plt.xlabel("Total Chronic Conditions")
plt.ylabel("Billing Amount ($)")
plt.show()
```

```python
# Correlation Plot: Total Conditions vs. Length of Stay
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Total Conditions", y="Length of Stay", data=data, alpha=0.5)
slope, intercept, r_value, _, _ = linregress(data["Total Conditions"], data["Length of Stay"])
plt.plot(data["Total Conditions"], intercept + slope * data["Total Conditions"], color="red")
plt.title(f"Correlation: Total Conditions vs Length of Stay (r = {r_value:.2f})")
plt.xlabel("Total Chronic Conditions")
plt.ylabel("Length of Stay (Days)")
plt.show()
```
